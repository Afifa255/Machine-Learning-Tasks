# Task 1: Exploring and Visualizing a Simple Dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset('iris')

# Print shape, column names, and first 5 rows
print("Dataset Shape:", iris.shape)
print("Column Names:", list(iris.columns))
print("\nFirst 5 rows:")
print(iris.head())
```

```python
print("--- Summary Info ---")
iris.info()
print("\n--- Descriptive Statistics ---")
print(iris.describe())
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species', palette='Set1')
plt.title('Sepal Length vs Sepal Width by Species')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend(title='Species')
plt.grid(True)
plt.show()
```

```python
# Create histograms for all numeric columns to show value distributions
numeric_cols = iris.select_dtypes(include=['float64', 'int64']).columns
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten()

for i, col in enumerate(numeric_cols):
    sns.histplot(data=iris, x=col, hue='species', kde=True, ax=axes[i], element='step', stat='density')
    axes[i].set_title(f'Distribution of {col}')
    axes[i].grid(True)

plt.tight_layout()
plt.show()
```

```python
# Create box plots for all numeric columns grouped by species to identify outliers
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten()

for i, col in enumerate(numeric_cols):
    sns.boxplot(data=iris, x='species', y=col, ax=axes[i], palette='Set2')
    axes[i].set_title(f'Box Plot of {col} grouped by Species')
    axes[i].grid(True)

plt.tight_layout()
plt.show()
```

## Conclusion: Write 2-3 lines summarizing what the visualizations reveal

1. The scatter plot reveals clear clustering, where the *setosa* species is completely distinct from *versicolor* and *virginica*, having shorter sepal lengths and wider sepal widths.
2. The histograms and box plots show that petal length and petal width have bimodal distributions, making them excellent features for species differentiation, whereas sepal measurements have overlapping distributions.
3. The box plots also identify some outliers in the sepal width of the *virginica* species.
